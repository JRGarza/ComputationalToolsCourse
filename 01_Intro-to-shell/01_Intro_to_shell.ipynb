## 1.1 Basics


The basic definition to the subject we will approach is "**a Unix shell is a command-line interpreter or shell that provides a command line user interface for Unix-like operating systems.**"

The most common shell, **or terminal**, is **bash** (bourne again shell) as it is the difault interpreter in many Linux distribution and versions of MacOS.

### Verifying which interpreter the user has

Let's first verify which shell we use as an interpreter. To do that open a terminal and type the following command
```bash
echo $SHELL
```

then press enter. If the printed statement is

```bash
/bin/bash
```
then your interpreter is `bash`. Other options of for the interpreter are `csh`, `tcsh`, `zsh` to mention some. If you have other interpreter as the default and you want to change to `bash` then you can type the following command:

```bash
chsh -s /bin/bash
```

and the available shells can be displayed with the command


```bash
cat /etc/shells
```

### Basic commands

As the terminal is an interpreter to control the operative system we need to learn some commands to interact with the system.



### $\bullet$ Where I am
Let's first open a terminal and verify which is the default direction, or `path` (this means which is the default folder to open when lunching a terminal). 
Let's type
```bash
pwd 
```

`pwd` prints the full path of the current location of the terminal. As default your terminal should open on your home directory, e.g. `/Users/jaime` in my case as my user is `jaime`.

### $\bullet$ Create new folders
To create a folder we use the command `mkdir`, short for"make directory", as
```bash
mkdir name_of_your_folder
```
Let's create a directory called `Astro`

### $\bullet$ Move to another place
If we want to access the directory that we just created, i.e. `Astro`, then we have to use the command `cd`, short for "change directory".


```bash
cd Astro
```

We cann use `cd` to move to any other folder if we know the `path` of the forlder, for example I will use an absolute `path` to move to another folder
```bash
cd /Users/jaime/Documents/UNIGE_master/03_Fall_2020/Python_course
```

we can go backwards to another folder using `..`, for example if I am inside `Astro` I will be able to go back to my home if I type 
```bash
cd ..
```
And to return to the last folder we visited we can use the `-` option. Type the following
```bash
cd -
pwd
```

We can go to our home folder from wherever we are by typing just
```bash
cd
```


### $\bullet$ ... and by the 8th day God created an ASCII file

We can create not only folders but empty files with the command `touch`, somethimes is is useful. To create an empty `.txt` file with the name `Newton` let's go **inside `Astro` folder** and type
```bash
touch Newton.txt
```

To see what is inside an ASCII file without editing it we use the command `less`
```bash
less Newton.txt
```

If you want to edit the file you need a text editor that can be called with bash, we will use `vim`. Let's type
```bash
vim Newton.txt
```
To start to edit we need to type `i` after opening the file. Let's write something inside the file:
```
ph'nglui mglw'nafh Cthulhu R'lyeh wgah'nagl fhtagn
```
To save the file we need to type `esc` key and then `:w` and enter. To save and exit let's type `esc` key and then `:wq` and enter. To exit without saving type `esc` key and then `:q!`. Using a text editor like `vim` is useful when tou are connecting to a remote server.

Lets type again 
```bash
less Newton.txt
```

### $\bullet$ I know where I am but do not know what else is here
When we are inside a folder we may want to know what is inside a folder, to do that let's type the comand `ls`, short for list, as
```bash
ls
```
To know more information about what is inside of the folder and hidden files let's type
```bash
ls -la
```



### $\bullet$ I know the name of the file but do not know where it is
To find if a file with a given name or pattern is inside a folder we need the `find` command
```bash
find . -name "name-of-the-file"
```

lets go to our home folder by typing only `cd`, now lets look for 
```bash
find . -name "Newton.txt"
```

I we remember the initial part of the name of the file but do not remember the whole name we can use `*` to tell the interpreter to fill the empty spaces, like  

to search find everithing that starts as `Newton.` we type
```bash
find . -name "Newton.*"
```

to search find all `.txt` files, let's type
```bash
find . -name "*.txt"
```

Such syntax is valid in any kind of shell command.

### $\bullet$ After creating, lets learn how to destroy
To remove a file (with no possible recovery) we use `rm`, short for remove, as
```bash
rm name-of-the-file-to-remove
```
If you want to remove a **folder** you will wive the option `-r` as
```bash
rm -r name-of-the-folder-to-remove
```

### $\bullet$ I forgot how to use a command
If you want to have more information about a command and all its options available thenwe use `man`, short for manual, as
```bash
man command-that-you-do-not-remember-how-to-use
```

Let's display the manual of `vim` as 
```bash
man vim 
```
